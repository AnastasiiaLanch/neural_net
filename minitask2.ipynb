```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
credit_data = pd.read_csv('/Users/nastalancinskaa/Desktop/AI/ML_start/credit_data.csv')
```

```python
df_credit_data = pd.DataFrame(credit_data)
```

```python
df_credit_data.replace('no data', np.nan, inplace = True)
df_credit_data['recent_salary'] = df_credit_data['recent_salary'].astype(np.float64)
```

```python
df_credit_data.isnull().sum()
```

```python
df_credit_data['recent_salary'] = df_credit_data['recent_salary'].fillna((df_credit_data['recent_salary'].median()))
```

```python
df_credit_data.isnull().sum()
```

```python
df_credit_data.dtypes
```

```python
df_credit_data.drop(['client_id'],axis=1,inplace=True)
```

```python
df_credit_data
```

```python
col = []
col2 = []
for i in df_credit_data.columns[0:16].tolist():
    if df_credit_data.dtypes[i] == 'int64' or df_credit_data.dtypes[i] == 'float64':
        col.append(i)
#print(col)
```

```python
plt.figure(figsize=(13,6))
sns.boxplot(data=df_credit_data[col],width=0.6)
plt.xticks(rotation=90)
plt.title('All number features boxplot')
plt.grid(True, linestyle='--', alpha=0.3, zorder=0)
plt.xlabel('features')
plt.ylabel('value')
plt.show()
```

```python

plt.figure(figsize=(8, 5))
plt.boxplot([df_credit_data['client_id']], labels=['A'])
plt.title('Boxplot через matplotlib')
plt.ylabel('Значения')
plt.show()
```

```python
plt.figure(figsize=(18,11))

plt.subplot(3, 3, 1)
sns.boxplot(df_credit_data['age'],width=0.6)
plt.xticks(rotation=90)
plt.title('age')
plt.grid(True, linestyle='--', alpha=0.3, zorder=0)
plt.xlabel('feature')
plt.ylabel('value')

plt.subplot(3, 3, 2)
sns.boxplot(df_credit_data['number_of_children'],width=0.6)
plt.xticks(rotation=90)
plt.title('number_of_children')
plt.grid(True, linestyle='--', alpha=0.3, zorder=0)
plt.xlabel('feature')
plt.ylabel('value')

plt.subplot(3, 3, 3)
sns.boxplot(df_credit_data['number_of_properties'],width=0.6)
plt.xticks(rotation=90)
plt.title('number_of_properties')
plt.grid(True, linestyle='--', alpha=0.3, zorder=0)
plt.xlabel('feature')
plt.ylabel('value')

plt.subplot(3, 3, 4)
sns.boxplot(df_credit_data['value_of_properties'],width=0.6)
plt.title('value_of_properties')
plt.grid(True, linestyle='--', alpha=0.3, zorder=0)
plt.xlabel('feature')
plt.ylabel('value')

plt.subplot(3, 3, 5)
sns.boxplot(df_credit_data['outcome_amount'],width=0.6)
plt.title('outcome_amount')
plt.grid(True, linestyle='--', alpha=0.3, zorder=0)
plt.xlabel('feature')
plt.ylabel('value')

plt.subplot(3, 3, 6)
sns.boxplot(df_credit_data['time_worked_since_last_employment_mnth'],width=0.6)
plt.title('time_worked_since_last_employment_mnth')
plt.grid(True, linestyle='--', alpha=0.3, zorder=0)
plt.xlabel('feature')
plt.ylabel('value')

plt.tight_layout() 
plt.show()
```
